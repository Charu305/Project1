```python
# Sales by City
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming full_data_df1 is your original DataFrame
# Group by 'City' and sum the 'Order Number'
city_sales = full_data_df1.groupby('City')['Order Number'].sum().reset_index()

# Sort cities by Order Number and select top 10
city_sales = city_sales.sort_values(by='Order Number', ascending=False).head(10)

# Create the bar plot using Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(data=city_sales, x='City', y='Order Number', palette='viridis')

# Add title and labels
plt.title('Top 10 Cities by Orders', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Total Orders', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=10)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 4. Sales by Brand
Product_sales = full_data_df1.groupby('Brand')['Order Number'].sum().reset_index()

# Sort cities by Order Number and select top 10
Product_sales = Product_sales.sort_values(by='Order Number', ascending=False).head(10)

# Create the bar plot using Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(data=Product_sales, x='Brand', y='Order Number', palette='viridis')

# Add title and labels
plt.title('Top 10 Brand by Orders', fontsize=16)
plt.xlabel('Brand', fontsize=12)
plt.ylabel('Total Orders', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=10)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group by 'Product Name' and sum 'Order Number' to find the most and least popular products
product_popularity = full_data_df1.groupby('Product Name')['Order Number'].sum().reset_index()

# Sort products by 'Order Number'
product_popularity_sorted = product_popularity.sort_values(by='Order Number', ascending=False)

# Show the most and least popular products
most_popular = product_popularity_sorted.head(1)
least_popular = product_popularity_sorted.tail(1)

print(f"Most Popular Product: {most_popular}")
print(f"Least Popular Product: {least_popular}")
```

```python
# Add a 'Profit Margin' column (Unit Price - Unit Cost) / Unit Price
full_data_df1['Profit Margin'] = (full_data_df1['Unit Price USD'] - full_data_df1['Unit Cost USD']) / full_data_df1['Unit Price USD']

# Group by 'Product Name' to calculate average profit margin per product
profitability = full_data_df1.groupby('Product Name')['Profit Margin'].mean().reset_index()

# Sort products by Profit Margin
profitability_sorted = profitability.sort_values(by='Profit Margin', ascending=False)

# Display the top and bottom profitable products
most_profitable = profitability_sorted.head(1)
least_profitable = profitability_sorted.tail(1)

print(f"Most Profitable Product: {most_profitable}")
print(f"Least Profitable Product: {least_profitable}")
```

```python

# Merge the two DataFrames on the common column, e.g., 'Store Name' or 'Store ID'
merged_data = pd.merge(full_data_df1, Stores_df, on='StoreKey', how='left')

# Assuming 'Store Size (sq meters)', 'Open Date' are present in your data
store_performance = merged_data.groupby('StoreKey').agg({
    'Order Number': 'sum',
    'Square Meters': 'mean',
    'Open Date': 'min'
}).reset_index()

# Visualize total sales by store
plt.figure(figsize=(12, 8))
sns.barplot(data=store_performance, x='StoreKey', y='Order Number', palette='coolwarm')
plt.title('Store Performance by Sales')
plt.xlabel('Store')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by 'Store Location' (e.g., 'City' or 'Region') to calculate total sales by location
location_sales = merged_data.groupby('Country_y')['Order Number'].sum().reset_index()

# Visualize sales by geographical location (e.g., City)
plt.figure(figsize=(10, 6))
sns.barplot(data=location_sales, x='Country_y', y='Order Number', palette='coolwarm')
plt.title('Sales by Store Location')
plt.xlabel('Store Location')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
